# 读取文件

```python
titanic = pd.read_csv("data/titanic.csv")
print(titanic)
print(titanic.head(8))
print(titanic.dtypes)

# 将pd 对象转成 xlsx
titanic.to_excel("data/titanic.xlsx", sheet_name="test", index=False)


# 读取 xlsx 为 pd对象
titanic = pd.read_excel("data/titanic.xlsx",sheet_name="test")
titanic.info()
```

```python
thefirst=titanic["1"]
print(type(thefirst))
print(thefirst.shape) # 返回多少行多少列  (27,)  表示 27行1列

thefirst_second=titanic[["1","2"]]
print(type(thefirst_second))
print(thefirst_second.shape) # 返回多少行多少列  表示 27行2列
```

```python
# 某列大于3的 
above_3 = titanic[titanic["1"]>3]
above_3.shape
```

```python
class_23=titanic[titanic["4"].isin([8,9])]
print(class_23)

# isin  等价于   使用| 需要使用 括号进行括起来
class_23=titanic[(titanic["4"]== 8) | (titanic["4"]== 9)]
print(class_23)
                 
```

```python
# 非空获取

age_no_null = titanic[titanic["3"].notna()]
age_no_null.shape
```

```python
# 编号为3的列 大于8 ，取第4列
adult_names = titanic.loc[titanic["3"]>8,"4"]
print(adult_names.head())

# iloc[x:y,n:m]
#  x+1 行 到 y行
# n+1列到 m列
print(titanic.iloc[1:3,2:4]) # 输出 第2行到第3行，第3列到第4列
```

```python
# 设置某行的某列为aaa
titanic.iloc[0:3,3] = "aaa"
titanic.head()
```
